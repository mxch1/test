__A8.__ _(7 points)_ In this part, we are going to find the similarity of locations based on their races population distributions using _cosine similarity_. In cases where there is no population of one race for a location in the dataframe (the corresponding value is `NaN`), replace them with `zero` using the `.fillna()` method of `pandas`. Then, create a list and append to that each pair of locations with their similarity as a tuple, like: `(loc1, loc2, similarity value)`.


```python
import numpy as np #needed for array mulitplication and dot product
```

```python
#understanding similarity 
#https://skipperkongen.dk/2018/09/19/cosine-similarity-in-python/
#https://masongallo.github.io/machine/learning,/python/2016/07/29/cosine-similarity.html

# get the columns that 
race_columns= ['White', 'Black', 'Hispanic', 'Asian', 'American_Indian_or_Alaska_Native', \
                'Native_Hawaiian_or_Other_Pacific_Islander', 'Two_Or_More_Races']
print(race_columns)
```

```python
#get the rows of df 
df.shape[0]
```

```python
#get first array 
loc1 = np.array(df.loc[0,race_columns].fillna(0)) #use fillna to fix NaN values 
for i in loc1:
    print(i)
```

```python
#get second array 
loc2 = np.array(df.loc[1,race_columns].fillna(0)) #use fillna to fix NaN values 
loc2 
```

```python
#cosines do dot product 
dot1 = loc1.dot(loc2)
dot1
```

```python
#cosines normalize 
norm = np.linalg.norm(loc1) * np.linalg.norm(loc2)
norm
```

```python
#cosines similarity score 
sim = dot1 / norm 
sim
```

```python
#make a function to do this - but performance? Maybe better in-line 
def cos_calc(loc1,loc2,df):
        loc1_ratio = np.array(df.loc[loc1,race_columns].fillna(0))
        loc2_ratio = np.array(df.loc[loc2,race_columns].fillna(0))
        sim_val = loc1_ratio.dot(loc2_ratio) / ((np.linalg.norm(loc1_ratio) * np.linalg.norm(loc2_ratio)))
        return sim_val
```
